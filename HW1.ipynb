```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apples = pd.read_csv('apple_quality.csv')
```

```python
apples
```

```python
apples.drop(apples.tail(1).index,inplace=True)
```

```python
apples
```

```python
plt.hist(apples.Size)
```

```python
good_apples=apples[apples.Quality=='good']
```

```python
bad_apples=apples[apples.Quality=='bad']
```

```python
plt.hist(good_apples.Size, alpha=0.5, label='good')
plt.hist(bad_apples.Size, alpha=0.5, label='bad')
plt.xlabel("Size")
plt.ylabel("Count")
plt.legend(loc='upper right')
plt.show()
```

```python
plt.scatter(good_apples.Size, good_apples.Ripeness, label = 'good')
plt.scatter(bad_apples.Size, bad_apples.Ripeness, label = 'bad')
plt.xlabel("Size")
plt.ylabel("Ripeness")
plt.legend(loc='upper right')
```
